```python
df1= pd.read_csv(url1)
df2= pd.read_csv(url2)
df3= pd.read_csv(url3)
```

```python
def dataframe_info(df):
    print("Shape of the dataframe: ", df.shape)
    print("Number of rows and columns: ", df.size)
    print("Number of columns: ", len(df.columns))
    print("info",df.info())
    return 
```

```python
def trim(df):
    # Strip leading and trailing whitespace from column names
    df.columns = df.columns.str.strip()

    # Convert all column names to lowercase for consistency
    df.columns = df.columns.str.lower()

    # Replace spaces in column names with underscores for easier access
    df.columns = df.columns.str.replace(' ', '_')

    # Print confirmation message after cleaning
    print("All column names cleaned, duplicates dropped, and text columns processed.")

    # Return the cleaned DataFrame
    return df   
```

```python
trim(df1)
```

```python
trim(df2)
```

```python
trim(df3)
```

```python
def clean_inconsistent_values(df):
    # 1. Standardize Gender
    df['gender'] = df['gender'].str.strip().str.lower().map({
        'f': 'F', 'female': 'F', 'femal': 'F',
        'm': 'M', 'male': 'M'
    })

    # 2. Standardize State abbreviations
    state_mapping = {
        'AZ': 'Arizona',
        'Cali': 'California',
        'WA': 'Washington'
    }
    df['state'] = df['state'].replace(state_mapping)

    # 3. Fix Education values
    df['education'] = df['education'].replace({'Bachelors': 'Bachelor'})
    df['education'] = df['education'].replace({'High School or Below': 'High School'})
    # 4. Simplify Vehicle Class
    df['vehicle_class'] = df['vehicle_class'].replace({
        'Sports Car': 'Luxury',
        'Luxury SUV': 'Luxury',
        'Luxury Car': 'Luxury'
    })

    # 5. Remove % from Customer Lifetime Value
    df['customer_lifetime_value'] = df['customer_lifetime_value'].astype(str).str.replace('%', '', regex=False)

    return df
```

```python
clean_inconsistent_values(df1)
```

```python
clean_inconsistent_values(df2)
```

```python
clean_inconsistent_values(df3)
```

```python
concat_df=pd.concat([df1,df2,df3],axis=0)
```

```python
dataframe_info(concat_df)
```
